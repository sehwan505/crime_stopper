```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
```

```python
data = pd.read_json("stream_result.json", lines=True)
data
```

```python
data = data.drop(data.columns[data.isnull().sum()>0], axis=1)

data["user"][0]
```

```python
for i in ["screen_name","name","followers_count"]: #user디렉토리에 있는 name,screen_name등을 따로 컬럼으로 만듬
    temp = []
    for j in data["user"]:
        temp.append(j[i])
    data[i] = temp
```

```python
temp = [] #엔티티의 hashtags가져옴
for j in data["entities"]:
    temp.append(j["hashtags"])
data["hashtags"] = temp
```

```python
twitter_id1 = pd.read_csv("twitter_id.csv")
twitter_id2 = pd.read_csv("twitter_id2.csv")
```

```python
twitter_id1["name"].str.find("또또")
```

```python
-twitter_id2["name"].count()
```

```python
#전에 활동했던 사람들이 아직도 활동한는 지 확인

count = 0
for i in twitter_id1["name"]:
    if twitter_id2["name"].str.find(i).sum() > -(twitter_id2["name"].count()):
        print(f"{i}가 중복 됩니다.")
        count += 1
print(count)        
```

```python
#트위터 아이디 모음
twitter_id_all = pd.concat([twitter_id1,twitter_id2])
twitter_id_all = twitter_id_all.drop_duplicates(subset=["name"])
twitter_id_all = twitter_id_all.drop(['screen_name','id'], axis=1)
twitter_id_all['IsCrime'] = 1

data  = pd.merge(data,twitter_id_all, on="name", how="left")

data["IsCrime"]=data["IsCrime"].fillna(0)
```

```python
hashtag_list = [] #hashtags 딕셔너리에서 꺼냄
hashtag_all = []
for i in data["hashtags"]:
    temp = []
    if len(i) == 0:
        continue
    for j in range(len(i)):
        temp.append(i[j]['text'])
        hashtag_all.append(i[j]['text'])
    hashtag_list.append(temp)
```

```python
hashtag_list = pd.Series(hashtag_list)
data["hashtags"] = hashtag_list
```

```python
data.loc[data["hashtags"].isnull(),"hashtags"] = ""
```

```python
count={} #hashtag의 빈도수
lists = hashtag_all
for i in lists:
    try: count[i] += 1
    except: count[i]=1
count = sorted(count.items(), reverse=True, key=lambda item: item[1])
```
